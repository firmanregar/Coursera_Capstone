```python
X= preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python

Ks = 70
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat0=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat0)

    
    std_acc[n-1]=np.std(yhat0==y_test)/np.sqrt(yhat0.shape[0])

mean_acc
```

```python
k = 25
#Train Model
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
```

```python
yhat_1 = neigh.predict(X_test)
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat_1))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion="entropy", max_depth = 15)
tree # it shows the default parameters
tree.fit(X_train,y_train)
yhat_2 = tree.predict(X_test)
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat_2))
```

```python
from sklearn import svm
machine=svm.SVC(kernel='rbf')
machine.fit(X_train, y_train) 
yhat_3 = machine.predict(X_test)
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat_3))
```
